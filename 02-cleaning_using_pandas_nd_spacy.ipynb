```python
# Lets create a function to clean data at once
def lower_nd_replace(series):
    output = series.str.lower()                                # converting to lower
    output = series.str.replace(r'\[.*?\]','', regex = True)   # removed brackets and data inside it
    output = series.str.replace(r'[^\w\s]','', regex = True)   # removed all punctuations
    return output
```

```python
'''
Lemmatization:
    - It reduces words to their base form.
    - spaCy uses combination of linguistic rules and stats models to lemmatize text.
    e.g running => run

There is alternative for lemmatization i.e stemming.

whats Difference between lemmatization and stemming ?

They both reduce words to their base form but lemmatization is the smarter way to approach 
and generally perform better.

   stemming -> am => am,is => is , are=> ar, happy => happi, happiness => happi
   Lemmatization -> am => be,is => be , are=> be, happy => happy, happiness => happy
'''


# token.lemma_ => Gives lemmatization for token
# ['I', 'be', 'sell', 'lemon', 'for', '$', '5', 'today']
# Here in this case 'i' gets converted into 'I', "'m" gets conevrted to 'be' i.e to base form
# lemmatised_words = []

# for token in doc:
#     lemmatised_words.append(token.lemma_)

# print(lemmatised_words)


lemmatised_words = [token.lemma_ for token in doc]
print(lemmatised_words)
# "i'm selling lemons for $5 today"
```

```python
'''
Stop words Removal:
    They are words without any significant meaning.
    To check list of stopwords => print(nlp.Defaults.stop_words)
    Stop words are common words that appear very frequently in a language,
    but carry little meaningful information for many NLP tasks.
'''

filtered_words = []

for word in lemmatised_words:
    token = nlp(word)[0]   # convert word to spaCy token
    if not token.is_stop:
        filtered_words.append(word)

print(filtered_words)
# ['I', 'be', 'sell', 'lemon', 'for', '$', '5', 'today']
```

```python

# Function to make sentence lower and clean it.
def lower_nd_replace(series):
    output = series.str.lower()
    output = output.str.replace(r'\[.*?\]', '', regex=True)
    output = output.str.replace(r'[^\w\s]', '', regex=True)
    return output


# Lets create a function that will get a token , lemmas, removed stop word
import spacy
nlp = spacy.load('en_core_web_sm')
def token_lemma_remove_stopwords(text):
    doc = nlp(text)
    filtered_words = []
    lemmatised_words = [token.lemma_ for token in doc]
    for word in lemmatised_words:
        token = nlp(word)[0]   # convert word to spaCy token
        if not token.is_stop:
            filtered_words.append(word)
        
    output = " ".join(filtered_words)
    return output

# Filter by POS
def filter_pos(text, pos_list = ["NOUN","PROPN"]):
    doc = nlp(text)
    
    output = []

    for token in doc:
        if token.pos_ in pos_list:
            output.append(token.text)      # returns the nouns and pronouns
    
    output = " ".join(output)
    return output
```

```python
# Creating pipeline for NLP task
def nlp_pipeline(series):
    output = lower_nd_replace(series)
    output = output.apply(token_lemma_remove_stopwords)
    output = output.apply(filter_pos)
    return output
```
