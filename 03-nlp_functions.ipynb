```python

# Function to make sentence lower and clean it.
def lower_nd_replace(series):
    output = series.str.lower()
    output = output.str.replace(r'\[.*?\]', '', regex=True)
    output = output.str.replace(r'[^\w\s]', '', regex=True)
    return output


# Lets create a function that will get a token , lemmas, removed stop word
import spacy
nlp = spacy.load('en_core_web_sm')
def token_lemma_remove_stopwords(text):
    doc = nlp(text)
    filtered_words = []
    lemmatised_words = [token.lemma_ for token in doc]
    for word in lemmatised_words:
        token = nlp(word)[0]   # convert word to spaCy token
        if not token.is_stop:
            filtered_words.append(word)
        
    output = " ".join(filtered_words)
    return output

# Filter by POS
def filter_pos(text, pos_list = ["NOUN","PROPN"]):
    doc = nlp(text)
    
    output = []

    for token in doc:
        if token.pos_ in pos_list:
            output.append(token.text)      # returns the nouns and pronouns
    
    output = " ".join(output)
    return output


# Creating pipeline for NLP task
def nlp_pipeline(series):
    output = lower_nd_replace(series)
    output = output.apply(token_lemma_remove_stopwords)
    output = output.apply(filter_pos)
    return output

# from nlp_functions import nlp_pipeline
```
